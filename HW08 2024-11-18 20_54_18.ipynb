```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt

# Step 1: Prepare the data
data = {
    'Income': ['High', 'Medium', 'Low', 'Medium', 'High', 'Low', 'Medium'],
    'Credit Score': ['Good', 'Good', 'Average', 'Poor', 'Average', 'Good', 'Average'],
    'Loan Amount': ['Low', 'Low', 'High', 'Medium', 'Medium', 'Low', 'Medium'],
    'Loan Approval': ['Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Encode categorical variables
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_encoded = df.apply(encoder.fit_transform)

# Separate features and target
X = df_encoded[['Income', 'Credit Score', 'Loan Amount']]
y = df_encoded['Loan Approval']

# Step 2: Train the decision tree model
tree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
tree.fit(X, y)

# Step 3: Visualize the tree
plt.figure(figsize=(10, 6))
plot_tree(tree, feature_names=['Income', 'Credit Score', 'Loan Amount'], class_names=encoder.inverse_transform([0, 1]), filled=True)
plt.title("Decision Tree for Loan Approval")
plt.show()

# Step 4: Print textual representation of the tree
print(export_text(tree, feature_names=['Income', 'Credit Score', 'Loan Amount']))
```

1. Decision Trees:
	•	Overview: Decision trees are a non-linear model used for both classification and regression tasks. They work by splitting data based on feature values to make predictions.
	•	Classification (Binary or Multiclass):
	•	Binary Classification: In tasks where the target variable has two possible outcomes (e.g., Yes/No, 0/1), decision trees split data to predict the class with the majority of samples in each leaf node.
	•	Multiclass Classification: Decision trees can also handle more than two classes, predicting the most frequent class in each leaf.
	•	Regression: Decision trees can be used for regression tasks by predicting a continuous value. At each leaf, the tree outputs the average value of the target variable for the data points in that leaf.
	•	How It Works: Decision trees recursively split the data based on a feature (standard) and a threshold that minimizes a criterion like Gini impurity or entropy. This continues until a stopping condition is met (e.g., a pure node or a pre-set depth).

1. Understanding Decision Trees and Classification Models

	•	We discussed the core concepts of decision trees, such as Gini impurity, samples, value, and class.
	•	Gini Impurity measures how “mixed” or “impure” the samples at a node are. A Gini value of 0 indicates pure nodes (all samples belong to one class).
	•	Samples represent the total number of data points at a node.
	•	Value shows the count of samples in each class at a node.
	•	Class is the predicted class for a node, typically the majority class.
	•	These metrics help to determine how the tree splits the data and how decisions are made at each node.

2. Decision Trees vs. Linear Regression

	•	Decision Trees are often used for classification tasks, where the goal is to predict discrete categories (e.g., yes/no, spam/ham).
	•	Regression models like Linear Regression are typically used for predicting continuous values (e.g., prices, quantities).
	•	Key differences:
	•	Decision Trees handle non-linear data and can easily model complex relationships between features, while Linear Regression assumes a linear relationship.
	•	Decision Trees work well with categorical variables, whereas Linear Regression usually requires continuous variables.

3. Accuracy vs. Sensitivity, Specificity, and Precision

	•	We discussed various metrics:
	•	Accuracy: The proportion of correctly predicted instances (both true positives and true negatives) out of all instances.
	•	It’s most useful when the classes are balanced (similar number of positive and negative instances).
	•	Sensitivity (Recall): Measures the proportion of actual positives correctly identified. High sensitivity is important in cases where it’s critical to identify all positive cases (e.g., medical tests).
	•	Specificity: Measures the proportion of actual negatives correctly identified. It’s crucial in situations where false positives need to be minimized (e.g., avoiding misdiagnosis in medical tests or minimizing spam email errors).
	•	Example: Spam Email Filtering—High specificity reduces the chances of legitimate emails being incorrectly flagged as spam.
	•	Precision: Focuses on the proportion of positive predictions that are truly correct. It’s especially important when false positives are costly (e.g., fraud detection or medical diagnostics).
	•	Example: Medical Diagnostics—High precision ensures that positive test results (e.g., cancer diagnosis) are more likely to be accurate, reducing unnecessary treatments for false positives.

4. False Positives and Probability

	•	We discussed how false positives work and why events flagged as false positives are more likely to be true false positives in the future. This is because the model has learned specific patterns that lead to these errors, and it tends to repeat the same mistakes.
	•	False Positives occur when a model incorrectly classifies a negative instance as positive (e.g., a legitimate email marked as spam).
	•	Once an event is labeled a false positive, the system’s behavior increases the probability that similar events will also be falsely classified as positive, leading to repeated false positives.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the data
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Remove rows with missing values (as implied in previous code)
ab_reduced_noNaN = ab.dropna()

# Create the 80/20 train-test split
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(
    ab_reduced_noNaN, test_size=0.2, random_state=42)

# Report the number of observations in each set
train_size = ab_reduced_noNaN_train.shape[0]
test_size = ab_reduced_noNaN_test.shape[0]

print(f"Training set size: {train_size} observations")
print(f"Testing set size: {test_size} observations")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assuming ab_reduced_noNaN_train and ab_reduced_noNaN_test are already defined
# and you've prepared y and X as shown in the previous steps

# Create the feature set (X) and target variable (y) from the training set
X_train = ab_reduced_noNaN_train[['List Price']]
y_train = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']

# Create the decision tree classifier with max_depth=2
clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Now, to evaluate the model, prepare the test data
X_test = ab_reduced_noNaN_test[['List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], rounded=True, fontsize=12)
plt.show()
```
